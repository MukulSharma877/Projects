## Task 2

```python
# Import Libraries

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
# Read Datasets

movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
# merging datasets

data = pd.merge(rating, movies, on = "movieId")
data.head()
```

```python
# Preprocessing the merged dataset

tfv = TfidfVectorizer()
matrix = tfv.fit_transform(movies["genres"])

movie_user = data.pivot_table(index="userId", columns="title", values = "rating")
movie_user
```

```python
# To find the similar movies, we need to find the cosine similarity between the movies

cosine_sim = linear_kernel(matrix, matrix)
cosine_sim
```

```python
# Content-Based Recommender System

indices=pd.Series(movies.index,index=movies['title'])
titles=movies['title']

def recommendation(title, N):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:N + 1]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
m = input("Enter the Movie Name: ")
n = int(input("Enter the Number of Recommendations Required: "))

recommendation(m, n)
```

## Task 3

```python
# Import Libraries

import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Read Datasets

movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
# merging datasets

data = pd.merge(rating, movies, on = "movieId")
data = data.drop(data.loc[:, ["movieId", "timestamp", "genres"]], axis = 1)
data.head()
```

```python
# Preprocessing the merged dataset

movie_user = data.pivot_table(index="title", columns="userId", values = "rating")
movie_user = movie_user.fillna(0)
movie_user
```

```python
# Content-Based Recommender System

from sklearn.neighbors import NearestNeighbors

df1 = movie_user.copy()

def movie_recommender(user, num_neighbors, num_recommendation):
  
  number_neighbors = num_neighbors

  knn = NearestNeighbors(metric='cosine', algorithm='brute')
  knn.fit(movie_user.values)
  distances, indices = knn.kneighbors(movie_user.values, n_neighbors=number_neighbors)

  user_index = movie_user.columns.tolist().index(user)

  for m,t in list(enumerate(movie_user.index)):
    if movie_user.iloc[m, user_index] == 0:
      sim_movies = indices[m].tolist()
      movie_distances = distances[m].tolist()
    
      if m in sim_movies:
        id_movie = sim_movies.index(m)
        sim_movies.remove(m)
        movie_distances.pop(id_movie) 

      else:
        sim_movies = sim_movies[:num_neighbors-1]
        movie_distances = movie_distances[:num_neighbors-1]
           
      movie_similarity = [1-x for x in movie_distances]
      movie_similarity_copy = movie_similarity.copy()
      nominator = 0

      for s in range(0, len(movie_similarity)):
        if movie_user.iloc[sim_movies[s], user_index] == 0:
          if len(movie_similarity_copy) == (number_neighbors - 1):
            movie_similarity_copy.pop(s)
          
          else:
            movie_similarity_copy.pop(s-(len(movie_similarity)-len(movie_similarity_copy)))
            
        else:
          nominator = nominator + movie_similarity[s]*movie_user.iloc[sim_movies[s],user_index]
          
      if len(movie_similarity_copy) > 0:
        if sum(movie_similarity_copy) > 0:
          predicted_r = nominator/sum(movie_similarity_copy)
        
        else:
          predicted_r = 0

      else:
        predicted_r = 0
        
      df1.iloc[m,user_index] = predicted_r
  recommend_movies(user,num_recommendation)

def recommend_movies(user, num_recommended_movies):
  recommended_movies = []
  for m in movie_user[movie_user[user] == 0].index.tolist():
    index_df = movie_user.index.tolist().index(m)
    predicted_rating = df1.iloc[index_df, df1.columns.tolist().index(user)]
    recommended_movies.append((m, predicted_rating))

  sorted_rm = sorted(recommended_movies, key=lambda x:x[1], reverse=True)
  
  for recommended_movie in sorted_rm[:num_recommended_movies]:
    print(recommended_movie[0])
```

```python
u = int(input("Enter the UserId: "))
n = int(input("Enter the Number of Recommendations Required: "))
k = int(input("Enter the Threshold for similar users: "))

movie_recommender(u, k, n)
```
